```python
import sys,os
sys.path.append(os.path.abspath("../"))
%pylab notebook
from experiments import pico as pico
from experiments import litebird as litebird
import pysm
import healpy as h

def ensure_dir(file_path):
    directory = os.path.dirname(file_path)
    if not os.path.exists(directory):
        os.makedirs(directory)
        
respath="../results/"
ensure_dir(respath)

share_datapath="/scratch/nas_chluba/specdist/arotti/share_data/"
# share_datapath="../dataout/"
```

# Common settings

```python
nside=512
pix_idx=arange(h.nside2npix(nside))
sky_config = {
    'synchrotron' : pysm.nominal.models("s2", nside,pixel_indices=pix_idx),
    'dust' : pysm.nominal.models("d2", nside,pixel_indices=pix_idx),
    #'freefree' : pysm.nominal.models("f1", nside,pixel_indices=pix_idx),
    'cmb' : pysm.nominal.models("c1", nside,pixel_indices=pix_idx),
    #'ame' : models("a1", nside,pixel_indices=pix_idx),
}

sky = pysm.Sky(sky_config)
```

# PICO configuration

```python
datapath=share_datapath +  '/bmode_moments_sims/pico/s2d2/'
ensure_dir(datapath)

pico_config={
    'nside' : nside,
    'frequencies' : pico.channels, #Expected in GHz
    'use_smoothing' : True,
    'beams' : pico.beams, #Expected in arcmin
    'add_noise' : True,
    'sens_I' : pico.noiserms, 
    'sens_P' : pico.noiserms,
    'noise_seed' : 1234,
    'cmb_seed' : 1236,
    'use_bandpass' : False,
    'output_units' : 'uK_CMB',
    'output_directory' : datapath,
    'output_prefix' : fname_prefix,
}

pico_instrument = pysm.Instrument(pico_config)

pico_instrument.observe(sky)
```

# LiteBird Configuration

```python
datapath=share_datapath +  '/bmode_moments_sims/litebird/s2d2/'
ensure_dir(datapath)

litebird_config={
    'nside' : nside,
    'frequencies' : litebird.channels, #Expected in GHz
    'use_smoothing' : True,
    'beams' : litebird.beams, #Expected in arcmin
    'add_noise' : True,
    'sens_I' : litebird.noiserms, #Expected in units uK_RJ
    'sens_P' : litebird.noiserms,
    'noise_seed' : 1234,
    'cmb_seed' : 1236,
    'use_bandpass' : False,
    'output_units' : 'uK_CMB',
    'output_directory' : datapath,
    'output_prefix' : fname_prefix,
}

litebird_instrument = pysm.Instrument(litebird_config)

litebird_instrument.observe(sky)
```
